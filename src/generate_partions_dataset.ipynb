### Importar librerias

```python
import sys
from datetime import datetime
from pyspark.sql import SparkSession
from pyspark.sql.functions import * 
import pyspark.sql.functions as F
```

### Crear SparkSession

```python
spark = SparkSession.builder \
    .appName("Partitioning Data") \
    .config("spark.driver.memory", "8g") \
    .config("spark.executor.memory", "12g") \
    .config("spark.driver.memoryOverhead", "2g") \
    .config("spark.executor.memoryOverhead", "2g") \
    .config("spark.memory.offHeap.enabled", True) \
    .config("spark.memory.offHeap.size", "16g") \
    .config("spark.driver.maxResultSize", "2048m") \
    .getOrCreate()
```

### Cargo el dataset

```python
input_path = "data-amex/"
output_path = "data-amex-partitioned/"
```

```python
train = spark.read.option("inferSchema", "true").option("header", "true").parquet(input_path+"train.parquet")
train.show(5)
```

#### Particionar el dataset de train en base a una variable categórica

```python
train.write.partitionBy("B_30").parquet(output_path+"train.parquet")
```

#### Particionar el dataset de test en base a una variable categórica

```python
test = spark.read.option("inferSchema", "true").option("header", "true").parquet(input_path+"test.parquet")
test.write.partitionBy("B_30").parquet(output_path+"test.parquet")
```
